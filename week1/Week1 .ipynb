```python
i = 4
rows = i
cols = i

matrix = []                               # empty matrix

for r in range(rows):                     # for inner loop
    row = []                              # new empty row

    for c in range(cols):               # for outer loop
        row.append(2)                   # add 2 in row

    matrix.append(row)                   # add completed rows in matrix    


for row in matrix:                               # print matrix
    print('  '.join(map(str, row)))             # map(str , row)  = change numbers to string
                                                 #  (' '.join )   = add all values to space     like    0  0  0  0
```

```python

#   paramit loop in steric
#.......................................


for level in range (1, rows +1):
    print(' ' *( rows - level)+ '*' * (2 * level-1))
```

```python
i = 4
rows = i
cols = i

matrix = []                               # empty matrix

for r in range(rows):                     # for inner loop
    row = []                              # new empty row

    for c in range(cols):               # for outer loop
        row.append(0)                   # add 0 in row

    matrix.append(row)                   # add completed rows in matrix    


for row in matrix:                               # print matrix
    print('  '.join(map(str, row)))             # map(str , row)  = change numbers to string
              
```
