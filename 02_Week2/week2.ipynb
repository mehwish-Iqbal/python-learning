```python
class student:
    def __init__ (self, name , age, grade):                           # __init__         is constructer
        self.name = name                                             # self.             is instance veriable
        self.age = age
        self.grade = grade

    def display_info(self):
       print(f"student name: {self.name}")
       print(f"age:{self.age}")
       print(f"grade:{self.grade}")

    def is_eligible(self):
        if self.age >= 15:
            print (self.name,"is eligible for admission")
        else:
            print (self.name, "is not eligible for admission")
        
```

```python
                                                   # accessing object attributes & Methods
#                                              .................................................
#  creating objects                               
student1 = student("Ali", 16, "10th")
student2 = student("sara", 14, "8th")
```

```python
#  Accessing attributes
print(student1.name)
print(student2.grade)
```

```python
# calling method
student1.display_info()
student1.is_eligible()
```

```python
student2.display_info()
student2.is_eligible()
```

```python
class student:
    def __init__ ( self, name , grade):
        self.__grade = grade

    def get_grade(self):
        return self.__grade                 # private variable

student = student ("Ali", 90)
print (student.get_grade())
```

```python
class student :                                    # blueprint
   
    @property                                    
    def grade (self):
        return self.__grade
   
    def __init__(self,name, grade):
        self.name = name
        self.__grade = grade
        
    def get_grade(self):
        return self.__grade
   
    def set_grade(self,grade):
        if 0 <= grade <= 100:
            self.__grade = grade
        else:
            print("invalid grade!")


student = student("Ali", 90)
student.set_grade(95)
student. get_grade()       
```

```python
class  playlist:
    def __init__ (self, songs):
        self.songs = songs
   
    def __len__ (self):                            # defalt method 1
        return len (self.songs)
    
    def __add__ (self, other):                         # defolt method 2
        return playlist(self.songs + other.songs)
        
    
```

```python
p1 = playlist(["song1", "song2"])
```

```python
p2 = playlist(["song3"])
```

```python
print(len(p1))
```

```python
print(len(p1+p2))
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):                                  # defalt method 1
        return f"({self.x}, {self.y})"

    def __add__(self, other):                           # defalt method 2
        if not isinstance(other, Point):
            return NotImplemented

        return Point(self.x + other.x, self.y + other.y)


```

```python
p1 = Point(2, 3)
p2 = Point(4, 5)

print(p1 + p2)
```

```python
#                                         using Try , except , else & finally Blocks
#                                     ................................................
```

```python
# if we can divide  10/0: (ZeroDivisionError )willl show
# if we can divide  10/k or any letter:  (ValueError) will show  , so to mannage these errors we follow these steps.

num = int(input("enter a No:"))   
print(10/num)
```

```python
try:
    num = int (input("Enter a NO : "))
    print(10 / num)

except ZeroDivisionError:
    print("you can't divide by Zero!")

except ValueError:
    print("that's not a number!")

except Exception  as e:
    print(f"An unexpected error occurred: {e}")
```

```python
try:
    print("running code....")
except:
    print("error found!")
else:
    print("no errors found!")
finally:
    print("done!")
    
```
